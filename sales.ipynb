```python
# Convert date columns to datetime
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

**Compute total sales**

```python
total_sales=df['Sales'].sum()
total_sales
```

**Analyze Sales Trends Over Time**

```python
# Extract year and month from order date
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# Group by YearMonth and calculate total sales
sales_trends = df.groupby('YearMonth')['Sales'].sum().reset_index()

print(sales_trends)
```

```python
# Group by product and calculate total sales and total quantity
best_selling_products = df.groupby('Product Name').agg({'Sales': 'sum', 'Quantity': 'sum'}).reset_index()

# Sort by sales in descending order
best_selling_products = best_selling_products.sort_values(by='Sales', ascending=False)
best_selling_products.head(10)
```

```python
# 1.Sales trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.head(100), x='Order Date', y='Sales')
plt.title('Sales Trends Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
#2. Sales distribution by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title('Sales Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#3.Top 10 best-selling products by sales
plt.figure(figsize=(14, 7))
top_10_products_sales = best_selling_products.head(10)
sns.barplot(x='Sales', y='Product Name', data=top_10_products_sales, palette='viridis')
plt.title('Top 10 Best-Selling Products by Sales')
plt.show()
```

```python
#4. Sales vs. Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs. Profit')
plt.show()
```
